```python
def plot_histogram(df,column_name):
    plt.figure(figsize=(5,3))
    sns.histplot(df[column_name],kde = True)
    plt.title(f"Distrubution of {column_name}")
    
    #calculate the mean values for the columns
    col_mean =df[column_name].mean()
    col_median =df[column_name].median()

    #for vertical lines for mean and median
    plt.axvline(col_mean, color="red",linestyle="--",label="Mean")
    plt.axvline(col_median, color="blue",linestyle="-",label="Median")

    plt.legend()

    plt.show()

```

Data preprocessing

```python
df.head()
```

label encoding the targert col = churn col 
coverting no = 0 and yes = 1

```python
df["Churn"] = df["Churn"].replace({"Yes":1,"No":0})
```

```python
df["Churn"].value_counts()
```

Label encoding for rest categorical cols

here we are identifying columns with data types to perform further operations


```python
object_columns = df.select_dtypes(include="object").columns
```

```python
object_columns
```

```python
encoders = {}

for column in object_columns:
    label_encoder = LabelEncoder()
    df[column] = label_encoder.fit_transform(df[column])
    encoders[column] = label_encoder


with open("encoder.pkl",'wb') as f:
    pickle.dump(encoders,f)
```

## now we are training and testing

```python
x=df.drop(columns=["Churn"])
y = df["Churn"]
```

```python
x
```

```python
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y, test_size=0.2,random_state=42)
```

```python
print(y_train.shape)
```

```python
print(y_train.value_counts())
```

to balance the train data we use smote 

Synthetic Minority Oversampling Technique (SMOTE)

```python
smote = SMOTE(random_state=42)
```

```python
x_train_smote, y_train_smote = smote.fit_resample(x_train, y_train)
```

```python
print(y_train_smote.shape)
```

```python
print(y_train_smote.value_counts())
```

## Model Training


Training with default hyperparameters


```python
models = {
    "Decision Tree" : DecisionTreeClassifier(random_state=42),
    "Random Forest" : RandomForestClassifier(random_state=42),
    "XGBoost" : XGBClassifier(random_state = 42)
}
```

now we perform crossvalidation rather than accuracy score because it will give the accuracy on multiple folds

```python
cv_scores={}

for model_name,model in models.items():
    print(f"Training {model_name} with default parameters")
    scores = cross_val_score(model, x_train_smote, y_train_smote, cv=5,scoring = "accuracy")
    cv_scores[model_name] = scores
    print(f"{model_name} cross_validation_accuracy: {np.mean(scores):.2f}")
    print("-"*70)
```

```python
cv_scores
```

```python
rfc = RandomForestClassifier(random_state=42)
```

```python
rfc.fit(x_train_smote, y_train_smote)
```

```python
print(y_test.value_counts())
```

Model Evaluation

```python
y_test_pred = rfc.predict(x_test)

print("Accuracy Score:\n", accuracy_score(y_test, y_test_pred))
print("Confsuion Matrix:\n", confusion_matrix(y_test, y_test_pred))
print("Classification Report:\n", classification_report(y_test, y_test_pred))
```

```python
# save the trained model as a pickle file
model_data = {"model": rfc, "features_names": x.columns.tolist()}


with open("customer_churn_model.pkl", "wb") as f:
  pickle.dump(model_data, f)
```

 Load the saved  model and  build a Predictive System

 

```python
with open("customer_churn_model.pkl", "rb") as f:
  model_data = pickle.load(f)

loaded_model = model_data["model"]
feature_names = model_data["features_names"]
```

```python
print(loaded_model)
```

```python
print(feature_names)
```

```python
input_data = {
    'gender': 'Female',
    'SeniorCitizen': 0,
    'Partner': 'Yes',
    'Dependents': 'No',
    'tenure': 1,
    'PhoneService': 'No',
    'MultipleLines': 'No phone service',
    'InternetService': 'DSL',
    'OnlineSecurity': 'No',
    'OnlineBackup': 'Yes',
    'DeviceProtection': 'No',
    'TechSupport': 'No',
    'StreamingTV': 'No',
    'StreamingMovies': 'No',
    'Contract': 'Month-to-month',
    'PaperlessBilling': 'Yes',
    'PaymentMethod': 'Electronic check',
    'MonthlyCharges': 29.85,
    'TotalCharges': 29.85
}


input_data_df = pd.DataFrame([input_data])

with open("encoders.pkl", "rb") as f:
  encoders = pickle.load(f)


# encode categorical featires using teh saved encoders
for column, encoder in encoders.items():
  input_data_df[column] = encoder.transform(input_data_df[column])

# make a prediction
prediction = loaded_model.predict(input_data_df)
pred_prob = loaded_model.predict_proba(input_data_df)

print(prediction)

#result
print(f"Prediction: {'Churn' if prediction[0] == 1 else 'No Churn'}")
print(f"Prediciton Probability: {pred_prob}")
```
